```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
def plot(input: list, title: str):
    y_label = title

    plt.figure(figsize=(10, 5))
    plt.plot(np.arange(0, len(input)), input)
    plt.xlabel('Episode')
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()
```

```python
with open("./rewards/reward_per_round.pkl", "rb") as file:
    input = pickle.load(file)
    plot(input, "Rewards")
```

```python
with open("./rewards/total_rewards.pkl", "rb") as file:
    input = pickle.load(file)
    plot(input, "Rewards")
```
