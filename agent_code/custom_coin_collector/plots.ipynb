```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
file = open("./monitor_training/total_rewards.pkl", "rb")

input = pickle.load(file)
plt.figure(figsize=(10,6))
rounds = np.linspace(1,len(input), len(input))
plt.plot(rounds, input, '.')

# Make plot nice
plt.title('Total reward per training round', fontsize=15)
plt.xlabel('Round number', fontsize=12)
plt.ylabel('Total reward', fontsize=12)
plt.xlim(xmin=0)
plt.grid()
plt.savefig('./monitor_training/rewards.jpg', dpi=300)
```
