```python
file = open("./monitor_training/performanceEvaluation.pkl", "rb")

input = pickle.load(file)
plt.figure(figsize=(10,6))
rounds = np.linspace(1,len(input), len(input))
plt.plot(rounds, input, '.')

mean = round(np.mean(input),2)
std = round(np.std(input), 2)

maxY = max(input)

# Add hyerparameters to plot as text
hyper_params_text = "Mean = " + str(mean) + "    ;    Std = " + str(std)

# Get the current axes object
ax = plt.gca()
# Add the text at the top left corner (fraction coordinates)
ax.text(0.33, 0.9, hyper_params_text, fontsize=14,
        transform=ax.transAxes, va='top', ha='left')

# Make plot nice
plt.title('Performance evaluation', fontsize=17)
plt.xlabel('Round number', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.xlim(xmin=0)
plt.ylim(ymin=0)
plt.ylim(ymax=maxY*1.3)
plt.grid()
plt.savefig('./monitor_training/performance_evaluation.jpg', dpi=300)
```

```python
def plot_loss_alt(loss_list, num_points=100):
    l = np.array(loss_list)
    max_entries = int(len(l)/num_points)
    g = np.mean(l[:max_entries*num_points].reshape(num_points, -1), axis=1)
    x = np.arange(len(g))
    plt.plot(x, g)
    plt.yscale("log")
    plt.xlabel("step")
    plt.ylabel("loss")
    plt.title("DQN with Target Network")
    plt.tight_layout()
    plt.show()
    
def plot_loss(loss_list):
    t = np.arange(len(loss_list))
    l = np.array(loss_list)
    plt.plot(t, l)
    plt.yscale("log")
    plt.xlabel("step")
    plt.ylabel("loss")
    plt.title("DQN with Target Network")
    plt.show()
```

```python
loss = np.loadtxt("loss_log.txt")
plot_loss(loss)
```
