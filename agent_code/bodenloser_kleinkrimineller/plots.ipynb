```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
file = open("./monitor_training/hyperparameters.pkl", "rb")

hyperparameters = pickle.load(file)

learning_rate = hyperparameters[0]
discount_factor = hyperparameters[1]
decay_active = hyperparameters[2]
epsilon_decay = hyperparameters[3]
epsilon_decay_after_rounds = hyperparameters[4]
n_step = hyperparameters[5]
```

```python
file = open("./monitor_training/total_rewards.pkl", "rb")

input = pickle.load(file)
plt.figure(figsize=(10,6))
rounds = np.linspace(1,len(input), len(input))
plt.plot(rounds, input, '.')

# Make plot nice
plt.title('Total reward', fontsize=17)
plt.xlabel('Round number', fontsize=12)
plt.ylabel('Total reward', fontsize=12)
plt.xlim(xmin=0)
plt.grid()

# Add hyerparameters to plot as text
hyper_params_text = "Learning rate: " + str(learning_rate) + "\n" + \
					"Discount factor: " + str(discount_factor) + "\n" + \
					"Decay Active: " + str(decay_active) + "\n" + \
					"Epsilon decay: " + str(epsilon_decay) + "\n" + \
					"Decay every " + str(epsilon_decay_after_rounds) + " rounds" + "\n" + \
					"N-step " + str(n_step)

# Get the current axes object
ax = plt.gca()
# Add the text at the top left corner (fraction coordinates)
ax.text(0.5, 0.5, hyper_params_text, fontsize=14,
        transform=ax.transAxes, va='top', ha='left')

plt.savefig('./monitor_training/rewards.jpg', dpi=300)
```
